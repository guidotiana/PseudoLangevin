```python
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 12})

import sys
sys.path.append("../shared_code")
from utils.general import load_stuff, load_inputs, get_files
```

```python
dirlist = [
    "results/sim0",
    "results/sim1",
    "results/sim2",
    #"results/sim4",
    #"results/sim5",
]
start, end = 0, 100

ob_names = ['cost', 'metric', 'mod2', 'q', 'dK']
ob_labels = [r'$\mathcal{C}$', r'$1-\epsilon$', r'$|\theta|^2$', r'$q$', r'$\delta K$']

fig, axs = plt.subplots(figsize=(9, 5*len(ob_names)), nrows=len(ob_names), ncols=1)
for iax, (ax, ob_name, ob_label) in enumerate(zip(axs, ob_names, ob_labels)):
    for d in dirlist:
        data = load_stuff(f'{d}/data.dat')
        pars = load_inputs(f'{d}/pars.txt')
        label = f'T={pars["T_list"][0]}, ' + r'T$_{ratio}$' + f'={pars["Tratio_list"][0]:.1e}, ' + r'm$_{1}$' + f'={pars["m1_list"][0]:.1e}'
        ax.plot((data['move']*pars['dt_list'][0])[start:end], data[ob_name][start:end], label=label)
    
    if ob_name in ('metric', 'q'):
        ax.set_ylim([-0.02, 1.02])
    elif ob_name in ['mod2']:
        ax.set_yscale('log')
    elif ob_name in ['dK']:
        ax.axhline([0.], linestyle='--', color='gray')
        
    ax.set_ylabel(ob_label)
    ax.set_xlabel(r'$t_{s}$')
    ax.legend();
```

---
#### sim0

```python
d = 'results/sim0'
pars_files = get_files(f"{d}/weights", prefix="pars_")

ob_names = ['T_mb', 'std']
ob_labels = [r'T$_{mb}$', r'$\sigma_{mb}$']

x, y = {}, {}
for ob_name in ob_names:
    y[ob_name] = []
    for fn in pars_files:
        pars = torch.load(fn)
        y[ob_name].append(
            [v for k, v in pars[ob_name].items()]
        )
    y[ob_name] = np.array(y[ob_name]).T
    x[ob_name] = np.arange(len(pars_files))


fig, axs = plt.subplots(figsize=(9, 5*len(ob_names)), nrows=len(ob_names), ncols=1)
for ax, ob_name, ob_label in zip(axs, ob_names, ob_labels):
    for ylayer in y[ob_name]:
        ax.plot(x[ob_name], ylayer, linestyle='--')
        ax.scatter(x[ob_name], ylayer)
    if ob_name == 'T_mb':
        ax.axhline([pars['T']], linestyle='--', color='r')
    
    ax.set_ylabel(ob_label)
    ax.set_xlabel('file number')
    ax.set_yscale('log')
```
